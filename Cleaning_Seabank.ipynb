```python
# Memuat data dari file
df = pd.read_csv('/content/drive/MyDrive/Sentimen Analisis 3 Top Bank Digital/Seabank.csv')
```

```python
# Membaca dataset CSV

# Mengecek baris yang mengandung link di kolom 'content'
rows_with_links = df[df['content'].str.contains(r'http\S+|www.\S+', na=False)]

# Mengecek baris yang mengandung simbol di kolom 'content'
rows_with_symbols = df[df['content'].str.contains(r'[^\w\s]', na=False)]

# Mengecek baris yang mengandung whitespaces di kolom 'content'
rows_with_whitespaces = df[df['content'].str.contains(r'\s+', na=False)]

# Mengecek baris yang mengandung angka di kolom 'content'
rows_with_numbers = df[df['content'].str.contains(r'\d+', na=False)]

# Menampilkan hasil
print("Baris yang mengandung link di kolom 'konten_ulasan':")
print(rows_with_links[['content']])
print(f"Jumlah baris yang mengandung link: {len(rows_with_links)}\n")

print("Baris yang mengandung simbol di kolom 'content':")
print(rows_with_symbols[['content']])
print(f"Jumlah baris yang mengandung simbol: {len(rows_with_symbols)}\n")

print("Baris yang mengandung whitespaces di kolom 'content':")
print(rows_with_whitespaces[['content']])
print(f"Jumlah baris yang mengandung whitespaces: {len(rows_with_whitespaces)}\n")

print("Baris yang mengandung angka di kolom 'content':")
print(rows_with_numbers[['content']])
print(f"Jumlah baris yang mengandung angka: {len(rows_with_numbers)}\n")

# Menampilkan data dengan indeks 2494, jika ada
def check_and_print(index, df):
    if index in df.index:
        print(f"Data pada indeks {index}:")
        print(df.loc[index, 'content'])
    else:
        print(f"Data pada indeks {index} tidak ditemukan atau tidak memenuhi kondisi.\n")
```

```python
df.info()  # Informasi tentang tipe data dan jumlah nilai non-null dalam setiap kolom
```

```python
# Menghapus kolom yang tidak diperlukan untuk analisis
df = df.drop(columns=[
    'reviewId', 'userName', 'userImage', 'score', 'thumbsUpCount', 'reviewCreatedVersion', 'at', 'replyContent', 'repliedAt', 'appVersion'
])
df.head()
```

```python
import re
# Menghapus karakter '@' dari content_remove_link
df['content_remove_at'] = df['content'].apply(lambda x: re.sub(r'@', '', str(x)))
df
```

```python
# Menghapus link dari data content
df['content_remove_link'] = df['content_remove_at'].apply(lambda x: re.sub(r'http\S+|www.\S+', '', str(x)))
df
```

```python
import re
df['content_remove_simbol'] = df['content_remove_link'].apply(lambda x: re.sub(r'[^\w\s.,!?]', '', x))
df
```

```python
# Menghapus data angka dari kolom 'content_remove_simbol' dan menyimpannya ke kolom 'content_remove_angka'
df['content_remove_angka'] = df['content_remove_simbol'].apply(lambda x: re.sub(r'\d+', '', x))

df
```

```python
def remove_extra_spaces(text):
    return re.sub(r'\s+', ' ', text)

df['content_remove_spaces'] = df['content_remove_angka'].apply(remove_extra_spaces)
df
```

```python
print(df.duplicated().sum())  # Menampilkan jumlah data duplikat
```

```python
# Mengatasi duplikasi
df = df.drop_duplicates()
```

```python
print(df.duplicated().sum())  # Menampilkan jumlah data duplikat
```

```python
# Validasi data
df  # Memeriksa data setelah tahap pembersihan
```

```python
# Ambil hanya kolom 'content_remove_spaces'
df = df[['content_remove_spaces']]
df
```

```python
# Mengganti nama kolom untuk konten yang telah dibersihkan
df = df.rename(columns={'content_remove_spaces': 'content'})
```
